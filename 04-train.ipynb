```python
import numpy as np
import tensorflow as tf
from main_funcs import train, iotest, inference

class MY_FLAGS(DGCNN_FLAGS):

    # flags for model
    NUM_CLASS  = 4
    MODEL_NAME = "dgcnn"
    TRAIN      = True
    KVALUE     = 20
    DEBUG      = True
    EDGE_CONV_LAYERS  = 3
    EDGE_CONV_FILTERS = 64
    FC_LAYERS  = 2
    FC_FILTERS = [512,256]
    
    # flags for train/inference
    SEED           = 44
    LEARNING_RATE  = 0.001
    GPUS           = [0,1,2,3]
    MINIBATCH_SIZE = 6
    WEIGHT_PREFIX  = './weights/snapshot'
    KVALUE         = 20
    NUM_POINT      = 2048
    NUM_CHANNEL    = -1
    ITERATION      = 10000
    REPORT_STEP    = 10
    SUMMARY_STEP     = 50
    CHECKPOINT_STEP  = 500
    CHECKPOINT_NUM   = 10
    CHECKPOINT_HOUR = 0.4

    # flags for IO
    IO_TYPE    = 'h5'
    INPUT_FILE = ['/gpfs/slac/cryo/fs1/g/tomo-seg/uresnet-tomo-seg/mut05-training.hdf',]
    OUTPUT_FILE = ''
    BATCH_SIZE = 24
    LOG_DIR    = './logs'
    MODEL_PATH = ''
    DATA_KEY   = 'raw'
    LABEL_KEY  = 'onehot'
    SHUFFLE    = 1

    def parse_args(self):
        print("\n\n-- CONFIG --")
        for name in filter(lambda a: not a.startswith('__') and not callable( getattr(self,a) ), dir(self)):
            attribute = getattr(self,name)
            if type(attribute) == type(self.parser): continue
            print("%s\t= %r" % (name, getattr(self, name)))
            
        # Set random seed for reproducibility
        np.random.seed(self.SEED)
        tf.set_random_seed(self.SEED)

flags = MY_FLAGS()
flags.parse_args()
```
